#### a) Wizualizacja

Zwizualizuj funkcje gęstości prawdopodobieństwa następujących rozkładów teoretycznych:

gdzie:
- $N(\mu, \sigma)$ - rozkład normalny o wartości oczekiwanej $\mu$ i odchyleniu standardowym $\sigma$,
- $t(df)$ - rozkład t-Studenta o $df$ stopniach swobody,
- $\chi^2(df)$ - rozkład chi-kwadrat o $df$ stopniach swobody,

Zestaw wykresy na pojedynczej figurze (siatka wykresów 2x3). Przygotuj figurę zgodnie z wytycznymi z nagłówka notatnika. Dodatkowo:

```python
# YOUR CODE HERE
N = [(0,1), (0,0.75), (0,1.25), (-2,1), (2,1)]
T = [1, 2, 5, 30]
X = [1, 2, 3, 5, 8]
F1 = [(10,1), (10,2), (10,5), (10,10), (10,20)]
F2 = [(1,10), (2,10), (5,10), (10,10), (20,10)]
F3 = [(1,1), (2,2), (5,5), (25,25), (50,50)]

title = ["Normal distribution", "T-student distribution", "Chi-2 distribution", 
         "F Snedocer distribution (1)", "F Snedocer distribution (2)", "F Snedocer distribution (3)"]

colors = ['r', 'g', 'b', 'c', 'm', 'y']

fig, ax = plt.subplots(2,3,figsize=(15,10), facecolor='white')
fig.suptitle("Different distributions")

x1=np.linspace(-6,6,100)
x2=np.linspace(-0.01,2,100)
#rozkład normalny 
for i, elem in enumerate(N):
    sns.lineplot(x=x1 ,y=sp.stats.norm.pdf(x1, elem[0], elem[1]), color=colors[i], ax=ax[0,0], label=f"N({elem[0]}, {elem[1]})")
ax[0,0].set_xlim([-5, 5])
ax[0,0].set_ylim([0, 0.65])
#rozkład t-Studenta o 
sns.lineplot(x=x1, y=sp.stats.norm.pdf(x1, 0,1), color=colors[0], ax=ax[0,1], label="N(0,1)") # dodanie N(0,1)
for i, elem in enumerate(T):
    sns.lineplot(x=x1, y=sp.stats.t.pdf(x1, elem), color=colors[i+1], ax=ax[0,1], label=f"t({elem})")
ax[0,1].set_xlim([-5, 5])
ax[0,1].set_ylim([0, 0.65])

#rozkład chi-kwadrat 
for i, elem in enumerate(X):
    sns.lineplot(x=x2, y=sp.stats.chi2.pdf(x2, elem), color=colors[i], ax=ax[0,2], label=f"Chi2({elem})")
ax[0,2].set_xlim([0, 2])
ax[0,2].set_ylim([-0.01, 4])

#rozkład F Snedecora
for i, elem in enumerate(F1):
    sns.lineplot(x=x2, y=sp.stats.f.pdf(x2, elem[0], elem[1]), color=colors[i], ax=ax[1,0], label=f"Snedecor{elem}")
for i, elem in enumerate(F2):
    sns.lineplot(x=x2, y=sp.stats.f.pdf(x2, elem[0], elem[1]), color=colors[i], ax=ax[1,1], label=f"Snedecor{elem}")
for i, elem in enumerate(F3):
    sns.lineplot(x=x2, y=sp.stats.f.pdf(x2, elem[0], elem[1]), color=colors[i], ax=ax[1,2], label=f"Snedecor{elem}")

for i in range (3):
    ax[0,i].set_facecolor("white")
    ax[0,i].set_title(title[i+3])
    ax[0,i].set_xlabel("Value")
    ax[0,i].set_ylabel("Probability density")
    ax[0,i].legend(loc = "upper right", title="Distribution")


for i in range (3):
    ax[1,i].set_xlim([0, 2])
    ax[1,i].set_ylim([-0.01, 4])
    ax[1,i].set_facecolor("white")
    ax[1,i].set_title(title[i+3])
    ax[1,i].set_xlabel("Value")
    ax[1,i].set_ylabel("Probability density")
    ax[1,i].legend(loc = "upper right", title="Distribution")
```

#### b) Charakterystyka rozkładów
Na podstawie przygotowanych wykresów scharakteryzuje opisane rozkłady teoretyczne i odpowiedz na pytanie w jaki sposób zmiana wartości parametrów rozkładów wpływa na położenie i kształt ich funkcji gęstości prawdopodobieństwa.


Rozkład normalny — posiada charakterystyczny dla siebie kształt (kształt dzwonowaty), zmieniając średnią, zmieniamy jego położenie względem osi x. Wykres jest symetryczny względem średniej. Na odchylenie standardowe wpływa na kształt wykresu, im większe std tym wykres robi się niższy i rozciąga się na boki. Ta dystrybucja ma najwyżej położony punkt dla wartości średni w porównaniu z resztą wykresów.

Rozkład t-studenta - posiada podobny kształt do rozkładu normalnego. Wolniej zbiega do bardzo małych wartości niż rozkład normalny. W przypadku wzrostu liczby stopni swobody upodabnia się on coraz bardziej do rozkładu normalnego. Jest symetryczny względem zera.

Rozkład chi-kwadrat — jest on prawostronnie asymetryczny, ma tylko wartości dodatnie (suma kwadratów jest nieujemna, więc jego specyfikacja zapewnia nam nie ujemność). Wraz ze wzrostem liczby stopni swobody, wykres się wypłaszcza.

Rozkład F-Snedecora - Również posiada tylko wartości dodatnie, jest prawostronnie asymetryczny. Wraz ze wzrostem obu stopni swobody, powoli dąży do rozkładu normalnego. Kiedy wzrasta tylko jeden stopień swobody (niezależnie, który) wykresy przesuwają się w prawą stronę i zmieniają się ich wysokości. Przy małej wartości pierwszego ze stopni swobody (wykresy czerwony i zielony), charakterystyka wykresu trochę odbiega od reszty. Widoczny jest gwałtowny spadek wartości, który po chwili się wypłasza

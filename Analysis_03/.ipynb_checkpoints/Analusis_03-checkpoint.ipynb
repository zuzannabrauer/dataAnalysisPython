Analiza ruchu na moście Fremont Bridge w Seattle wykazała pewne interesujące tendencje, które warto szczegółowo przeanalizować.

W miesiącach wiosennych i letnich odnotowano większy ruch na moście, zarówno po wschodniej, jak i po zachodniej stronie. Jest to prawdopodobnie spowodowane wzrostem temperatury, która sprzyja aktywności fizycznej, takiej jak jazda rowerem. Szczególnie dużo rowerzystów obserwuje się na początku kwietnia, co może być skorelowane z dniem wolnym od pracy z okazji Wielkanocy oraz zwiększoną ilością czasu wolnego.

Natomiast na początku lipca i w okolicach świąt Bożego Narodzenia zaobserwowano nagły spadek ilości rowerzystów. Może to być związane z wyja acje i zmniejszoną dostępnością dojazdu rowerowego w tych okresach.

Warto również zwrócić uwagę na różnice w intensywności ruchu między wschodnią a zachodnią stroną mostu. Pomimo że większość obszarów o większym zagęszczeniu populacyjnym znajduje się na wschodnim brzegu jeziora Lake Union, obserwuje się większe natężenie ruchu rowerowego po zachodniej stronie mostu. Może to być spowodowane mniej zatłoczonym terenem na zachodnim brzegu, co sprawia, że jest to atrakcyjniejsza trasa dla rowerzystów.

Interesującym obserwacją jest również fakt, że najwięcej rowerzystów można zaobserwować w dni powszednie, zwłaszcza we wtorek, środę i czwartek. Może to sugerować, że wielu ludzi korzysta z rowerów jako środka transportu do pracy w tych dniach. Szczególnie duży spadek ruchu zauważalny jest w piątki oraz w weekendy, co może być związane z mniejszą liczbą osób podróżujących do pracy i zwiększonym czasem wolnym.

Warto również zwrócić uwagę na szczyty ruchu, które w dni powszednie przypadały na godz (7.00-8.00)iny poranne i p(16.00-17.00)opołudniowe, a w weekendy na godziny(12.00-16.00) południowe. To sugeruje, że w dni robocze ludzie korzystają z rowerów głównie do dojazdu do pracy, podczas gdy w weekendy preferują bardziej rekreacyjne przejażdżki.

```python
# Sprawdzam podejrzane okresy czasu w kwietniu i lipcu:
def highlight_max(s):
    is_max = s == s.max()
    return ['background-color: red' if v else '' for v in is_max]

def highlight_min(s):
    is_min = s == s.min()
    return ['background-color: green' if v else '' for v in is_min]
   
daily_bike_traffic_data = pd.read_pickle("daily_bike_traffic_data.pkl")
weather_data = pd.read_pickle("weather_data.pkl")

april = daily_bike_traffic_data[daily_bike_traffic_data.index.month == 4] 
april = april[april.index.day < 10]
weather_april = weather_data.loc[april.index,:]
df1 = april.copy()
df1[weather_april.columns] = weather_april
styled_df1 = df1.style.apply(highlight_max, subset=['Total'], axis=0)
display(styled_df1)

july = daily_bike_traffic_data[daily_bike_traffic_data.index.month == 7]
july = july[july.index.day < 10]
weather_july = weather_data.loc[july.index,:]
df2 = july.copy()
df2[weather_july.columns] = weather_july
styled_df2 = df2.style.apply(highlight_min, subset=['Total'], axis=0)
display(styled_df2)
```

```python
bike_traffic_data = pd.read_pickle("bike_traffic_data.pkl")

april_07 = bike_traffic_data[(bike_traffic_data.index.day==7) & (bike_traffic_data.index.month==4)]
april_07.index = april_07.index.hour

july_03 = bike_traffic_data[(bike_traffic_data.index.day==3) & (bike_traffic_data.index.month==7)]
july_03.index = july_03.index.hour

def plot_bike_traffic(ax, data, date, color,ylim):
    ax.set_facecolor("white")
    ax.plot(data.index, data.Total, c=color)
    ax.set_xlim([0, 23])
    ax.set_ylim(ylim)
    ax.set_xlabel("Godzina dnia")
    ax.set_ylabel("Liczba rowerzystów (Total)")
    ax.set_title(f"Rowerzyści na Fremont Bridge - {date}")

fig, ax = plt.subplots(2,1, figsize=(8,8), facecolor="white")
plot_bike_traffic(ax[0], april_07, "7 kwietnia 2022", "red",[0, 700])
plot_bike_traffic(ax[1], july_03, "3 lipca 2022", "green",[0, 80])

fig.suptitle("Anomalie")
fig.tight_layout()
```

W przypadku pierwszej anomalii z dnia 7 kwietnia 2022 roku, możliwe jest, że została ona wywołana nagłym ociepleniem. Był to czwartek, co oznacza, że po dłuższym okresie słabszej pogody, ludzie mogli skorzystać z poprawy warunków atmosferycznych, aby wybrać się na przejażdżkę rowerową do pracy. Nagły wzrost temperatury sprawił, że pogoda stała się znacznie przyjemniejsza do jazdy rowerem, co spowodowało zwiększone natężenie ruchu, charakterystyczne dla dojazdu i powrotu z pracy w dzień roboczy.

W przypadku drugiej anomalii z dnia 3 lipca 2022 roku, sytuacja mogła być spowodowana nagłym spadkiem temperatury i deszczu. Był to niedzielny dzień przed świętem niepodległości. Ludzie mieli zapewne w planach kolejny dzień świętowania, a niepogoda nie sprzyjała wyjściu na przejażdżkę rowerową. Nagłe pogorszenie warunków atmosferycznych mogło zniechęcić do wyjścia na rower.ców.

Przez większą część roku temperatura była dodatnia, średnia wartość temperatury w Seattle w 2022 wynosiła ok 10°C. W czasie wakacyjnym pojawiały sie temperatury powyżej 30°C, na przełomie czerwca i lipca pojawia sie kilkudniowy brak danych odnośnie mierzonej temperatury, podobna sytuacja pod koniec listopada W okresie zimowym temperatura kilkukrotnie była ujemna, ale przez przzeważający czas nie spadała poniżej 0°C.Temperatura w Seattle wykazywała wyraźną sezonowość, z najcieplejszymi miesiącami w lipcu i sierpniu oraz najchłodniejszymi w styczniu i lutym.

W tym regionie występuje duże zachmurzenie, w okresie letnim jest ono zdecydownaie zmniejszone, jednak nadal odczuwalne. Przyczyniają się do tego m.in. wpływ zimnych prądów morskich z Pacyfiku oraz bliskość gór Kaskadowych. Wpływ ma również tzw. "marine layer" - warstwa chłodnego i wilgotnego powietrza unoszącego się nad powierzchnią morza. W miesiącach zimowych zachmurzenie wynosi ok 90%.

Opady w Seattle są częstym zjawiskiem, występują przez ponad połowe roku. Suma opadów wynosi ponad 1000mm/m^2 w skali roku. Największe opady występują w lutym oraz w okresie zimowym. Natomiast latem, choć opady są zazwyczaj rzadsze, nadal mogą występować, zwłaszcza w formie krótkich, intensywnych burz. Jest to związane z przemieszczaniem się frontów atmosferycznych oraz zmianami w cyrkulacji atmosferycznej.Mimo że opady deszczu są dość powszechne, Seattle nie jest jednym z najbardziej deszczowych miast w Stanach Zjednoczonych. Ze względu na częstotliwość opadów, deszcz stanowi ważny element klimatu Seattle i wpływa na życie mieszkańców oraz na krajobraz miasta.

Średnia prędkość wiatru w Seattle w 2022 wynosiła ok 5m/s.Końcówka roku zdecydowanie była najbardziej wietrznym okresem roku.Wartości wachają sie pomiędzy 1m/s, a 17 m/s. Nie widać tendencji wzrostowej w konkretnych miesiącach, wzrost wietrzności występuje pikami. Ogólnie rzecz biorąc, wietrzność w Seattle może być większa w niektórych obszarach, takich jak wybrzeże Pacyfiku lub obszary bardziej wystawione na działanie wiatru, np. obszary otwarte. Z drugiej strony są obszary osłonięte, np. doliny, które mogą doświadczać mniejszej wietrzności.

Klimat Seattle w 2022 roku był umiarkowany, z łagodnymi zimami i ciepłymi letami.
Zachmurzenie było umiarkowane, z większym zachmurzeniem w miesiącach zimowych i mniejszym w miesiącach letnich.
Prędkość wiatru była umiarkowana.

Istnieje pozytywna korelacja między temperaturą a liczbą rowerzystów. Wraz ze wzrostem temperatury liczba rowerzystów ma tendencję do wzrostu. Współczynniki korelacji dla temperatury minimalnej, średniej i maksymalnej są wszystkie dodatnie i mieszczą się w przedziale od 0,7 do 0,8. Współczynniki determinacji również wskazują, że temperatura wyjaśnia dużą część wariancji w liczbie rowerzystów.

Istnieje słaba negatywna korelacja między całkowitymi opadami a liczbą rowerzystów. Wraz ze wzrostem opadów liczba rowerzystów ma tendencję do niewielkiego,wolnego spadku. Współczynnik determinacjijest dość niski, więc sugeruje to, że opady wyjaśniają niewielką część wariancji w liczbie rowerzystów.

Istnieje słaba negatywna korelacja między opadami śniegu i głęgokością puchu śnieżnego na glebie a liczbą rowerzystów. Współczynniki korelacji dla obu czynników zbliżają się do -0,1/ -0,2 powoduje to prawie nie istniejącą kolrealcje czynników, a współczynniki determinacji są minimalne, co wskazuje, że warunki śniegowe praktycznie nie mają wpływu na liczbę rowerzystów.

Istnieje słaba negatywna korelacja między średnią prędkością wiatru a liczbą rowerzystów. Współczynnik korelacji oraz współczynnik determinacji nadal są niskie. Występuje słaba negatywna korelacja między maksymalną prędkością wiatru a liczbą rowerzystów. Warunku wiatrowe, szczególnie jego maksymalna prędkość mają lekki wpływ na ilość rowerzystów.

Istnieje negatywna korelacja między zachmurzeniem a liczbą rowerzystów, jest ona większa niż w przypadku śniegu i wiatru.Współczynnik determinacji jest większy niz w poprzednich przypadkach, nadal jednak nie mozemy uznać go za wiodący. Zachmurzenie ma nieco bardziej zauważalny wpływ na liczbę rowerzystów w porównaniu z innymi parametrami pogodowymi.

Temperatura ma najsilniejszą pozytywną korelację z liczbą rowerzystów, co wskazuje, że cieplejsza pogoda zachęca więcej osób do jazdy na rowerze.
Znaczącym  aspektem jest też występowanie zachmurzenia oraz opadów deszczu, wpływa to na chęć jazdy na rowerze. Opady śniegu mają niewielki wpływ na liczbę, rowerzytów.

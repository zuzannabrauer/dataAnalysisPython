Korzystając z funkcji `scipy.stats.linregress()` przygotuj tablicę `regression_data_1` zawierającą wartości współczynników równania regresji liniowej pomiędzy wartościami parametrów pogody (zmienna niezależna, objaśniająca), a liczbą rowerzystów przejeżdżających przez most danego dnia (zmienna zależna, objaśniana).

- Tablica powinna zawierać wyłącznie następujące elementy:</br>

    > Kolumny:</br>
    > 1 rzędu: Total, East, West</br>
    > 2 rzędu: a, b (ten sam zestaw kolumn dla każdego indeksu 1 rzędu)</br>

Zestaw wykresy na pojedynczej figurze (siatka wykresów 3x3). Przygotuj figurę zgodnie z wytycznymi z nagłówka notatnika. Dodatkowo:

```python
# Tabela
# YOUR CODE HERE
regression_data_1 = pd.DataFrame(index = weather_data.columns, columns = pd.MultiIndex.from_product([['Total', 'East', 'West'], ['a', 'b']]))

for bike_column_name in daily_bike_traffic_data.columns:
    for weather_column_name in weather_data.columns:
        temp = pd.concat([daily_bike_traffic_data[bike_column_name], weather_data[weather_column_name]], axis = 1).dropna()
        regression_data_1.loc[weather_column_name, (bike_column_name, 'a')] = sp.stats.linregress(temp[weather_column_name], temp[bike_column_name]).slope
        regression_data_1.loc[weather_column_name, (bike_column_name, 'b')] = sp.stats.linregress(temp[weather_column_name], temp[bike_column_name]).intercept

regression_data_1.style.format(precision = 3)
```

#### b) Regresja krzywoliniowa

Pobieżna analiza wykresów przygotowanych we wcześniejszym podpunkcie pokazuje, że w kilku przypadkach relacje między zmiennymi są bardziej skomplikowane i prosta regresja liniowa nie jest dobrym przybliżeniem obserwowanego trendu.

Korzystając z funkcji `scipy.optimize.curve_fit()` przygotuj tablicę `regression_data_2` zawierającą współczynniki równań regresji między zachmurzeniem, a liczbą rowerzystów przejeżdżających przez most danego dnia oraz między dziennymi opadami, a liczbą rowerzystów przejeżdżających przez most danego dnia. W obu przypadkach skorzystaj z następującego równania regresji:
$$ y = a e^{-bx} + c$$ 
gdzie: a, b, c - współczynniki równania dobierane metodą najmniejszych kwadratów.

W przypadku regresji między dziennymi opadami, a liczbą rowerzystów przejeżdżających przez most danego dnia ustaw parametry funkcji w taki sposób, aby wartość współczynnika c zawierała się w przedziale 90% - 110% wartości średniej liczby rowerzystów z 10 dni w roku z największą ilością opadów.

- Tablice powinna zawierać wyłącznie następujące elementy:</br>
    > Indeks: </br>
    > 1 rzędu: Sky, Precipitation</br>
    > 2 rzędu: Cloud cover, Total (sparowane pojedynczo z odpowiadającymi im indeksami 1 rzędu tak jak w nazwach kolumn tablicy `weather_data`)</br>    
    
    > Kolumny:</br>
    > 1 rzędu: Total, East, West</br>
    > 2 rzędu: a, b, c (ten sam zestaw kolumn dla każdego indeksu 1 rzędu)</br>
    
Wyświetl przygotowaną tablicę w taki sposób aby zawarte w tablicy liczby wyświetlały się z dokładnością 3 miejsc po przecinku.

Następnie przedstaw wyniki analizy w postaci wykresów korelacyjnych poszczególnych parametrów pogody z całkowitą liczbą rowerzystów przejeżdżających przes most danego dnia z nałożonymi na nie krzywymi regresji. Zestaw wykresy na pojedynczej figurze (siatka wykresów 1x2). Przygotuj figurę zgodnie z wytycznymi z nagłówka notatnika. Dodatkowo:
- ustaw różne od domyślnych i od siebie nawzajem kolory widocznych na wykresach punktów i linii,
- w tytule lub legendzie każdego z wykresów załącz równania regresji sformatowane tak, aby parametry równania były wyświetlone z dokładnością 3 miejsc po przecinku,
- w przypadku zastosowania legend należy je umiejscowić tak, aby znajdowały się w obrębie pól wykresów, ale nie zasłaniały znajdujących się na nich danych.

```python
# Obliczenia i tablica
regression_data_2 = pd.DataFrame(index = weather_data.columns, columns = pd.MultiIndex.from_product([['Total', 'East', 'West'], ['a', 'b', 'c']]))
regression_data_2 = regression_data_2.loc[['Precipitation', 'Sky'], :]

def func(x, a, b, c):
    return a * np.exp(-b * x) + c

for bike_column_name in daily_bike_traffic_data.columns:
    regression_data_2.loc['Sky', (bike_column_name, 'a')] = sp.optimize.curve_fit(func, weather_data['Sky'].values.flatten().T, daily_bike_traffic_data[bike_column_name].values.flatten())[0][0]
    regression_data_2.loc['Sky', (bike_column_name, 'b')] = sp.optimize.curve_fit(func, weather_data['Sky'].values.flatten().T, daily_bike_traffic_data[bike_column_name].values.flatten())[0][1]
    regression_data_2.loc['Sky', (bike_column_name, 'c')] = sp.optimize.curve_fit(func, weather_data['Sky'].values.flatten().T, daily_bike_traffic_data[bike_column_name].values.flatten())[0][2]

indexes = weather_data.nlargest(10, ('Precipitation', 'Total'))
nlargest = daily_bike_traffic_data.loc[indexes.index, :]
mean = nlargest.mean()
lower_limit = 0.9*mean
upper_limit = 1.1*mean

i = 0
for bike_column_name in daily_bike_traffic_data.columns:
    bounds = ([-np.inf, -np.inf, lower_limit[i]], [np.inf, np.inf, upper_limit[i]])
    temp = pd.concat([daily_bike_traffic_data[bike_column_name], weather_data[('Precipitation', 'Total')]], axis = 1).dropna()
    regression_data_2.loc['Precipitation', (bike_column_name, 'a')] = sp.optimize.curve_fit(func, temp[('Precipitation', 'Total')].values.flatten().T, temp[bike_column_name].values.flatten(), bounds = bounds)[0][0]
    regression_data_2.loc['Precipitation', (bike_column_name, 'b')] = sp.optimize.curve_fit(func, temp[('Precipitation', 'Total')].values.flatten().T, temp[bike_column_name].values.flatten(), bounds = bounds)[0][1]
    regression_data_2.loc['Precipitation', (bike_column_name, 'c')] = sp.optimize.curve_fit(func, temp[('Precipitation', 'Total')].values.flatten().T, temp[bike_column_name].values.flatten(), bounds = bounds)[0][2]
    i += 1

regression_data_2.style.format(precision = 3)
```

```python
### Komórka testowa

# Testy wybranych wartości z tablicy regression_data_2
assert np.isclose(regression_data_2.loc[("Precipitation", "Total"), ('Total', 'a')], 1897.3445607675376)
assert np.isclose(regression_data_2.loc[(("Sky", "Cloud cover")), ('East', 'b')], -3.5974646434134456)
assert np.isclose(regression_data_2.loc[("Precipitation", "Total"), ('Total', 'c')], 697.29)
```

```python
# Wykres
fig, ax = plt.subplots(1, 2, figsize = (12, 6), constrained_layout = True)

temp = pd.concat([daily_bike_traffic_data['Total'], weather_data[('Precipitation', 'Total')]], axis = 1).dropna()
sns.scatterplot(data = temp, x = ('Precipitation', 'Total'), y = 'Total', ax = ax[0], color = 'gray')
x = np.linspace(temp[('Precipitation', 'Total')].min(), temp[('Precipitation', 'Total')].max())
ax[0].plot(x, regression_data_2.loc[('Precipitation', 'Total'), 'Total'][0] * np.exp(-regression_data_2.loc[('Precipitation', 'Total'), 'Total'][1]*x) + regression_data_2.loc[('Precipitation', 'Total'), 'Total'][2], c="red")
ax[0].set_ylabel('Total cyclist')
ax[0].set_title('y = ' + str(round(regression_data_2.loc[('Precipitation', 'Total'), 'Total'][0], 3)) + '*exp(' + str(round(regression_data_2.loc[('Precipitation', 'Total'), 'Total'][1], 3)) + '*x) + ' + str(round(regression_data_2.loc[('Precipitation', 'Total'), 'Total'][2], 3)))

temp = pd.concat([daily_bike_traffic_data['Total'], weather_data[('Sky', 'Cloud cover')]], axis = 1).dropna()
sns.scatterplot(data = temp, x = ('Sky', 'Cloud cover'), y = 'Total', ax = ax[1], color = 'gray')
x = np.linspace(temp[('Sky', 'Cloud cover')].min(), temp[('Sky', 'Cloud cover')].max())
ax[1].plot(x, regression_data_2.loc[('Sky', 'Cloud cover'), 'Total'][0] * np.exp(-regression_data_2.loc[('Sky', 'Cloud cover'), 'Total'][1]*x) + regression_data_2.loc[('Sky', 'Cloud cover'), 'Total'][2], c="red")
ax[1].set_ylabel('Total cyclist')
ax[1].set_title('y = ' + str(round(regression_data_2.loc[('Sky', 'Cloud cover'), 'Total'][0], 3)) + '*exp(' + str(round(regression_data_2.loc[('Sky', 'Cloud cover'), 'Total'][1], 3)) + '*x) + ' + str(round(regression_data_2.loc[('Sky', 'Cloud cover'), 'Total'][2], 3)))

fig.suptitle('Curve regression analysis for individual weather parameters and the total number of cyclists');
```

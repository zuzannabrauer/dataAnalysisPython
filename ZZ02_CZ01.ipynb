Zaimportuj zawarte w plikach dane do notatnika i utwórz na ich podstawie tablicę `recession_data_1` wykonaną według następujących wytycznych:

- Tablica powinna zawierać wyłącznie następujące elementy:</br>
    Indeks: Country Code</br>
        Wiersze powinny być indeksowane trzyliterowymi kodami krajów.</br>
    Kolumny: Country Name, Region Name, Income Group, 2006, 2007, 2008, 2009, 2010, 2011 (w tej kolejności, lata w nagłówkach jako zmienne typu str)</br>
        Kolumna Income Group zawiera informację o stopniu zamożności państwa wg danych z 2008 roku.
        Kolumny 2006-2011 zawierają wartości wskaźnika GDP growth (annual %) dla poszczególnych lat.
- Zmień występujące w kolumnie Income Category kody (H, UM, LM, L) na pełne wyrażenia (High, Upper middle, Lower middle, Low).
- Tablica nie powinna zawierać wierszy, w których znajdują się wartości NaN (w dowolnej kolumnie) oraz wiersza dotyczącego Libii.

Zapisz tablicę do pliku recession_data_1.pkl.

Wyświetl przygotowaną tablicę w taki sposób aby widoczne było po 5 pierwszych i ostatnich wierszy, a zawarte w tablicy liczby (za wyjątkiem lat w nagłówkach kolumn) wyświetlały się z dokładnością do 2 miejsca po przecinku.

```python
## YOUR CODE HERE
rows_to_import_from_raw_data_3 = list(range(10,230))
data1 = pd.read_csv('raw_data_1.csv', nrows = 217)
data2 = pd.read_csv('raw_data_2.csv', delimiter = ';')
data3 = pd.read_excel('raw_data_3.xlsx', sheet_name = 'Country Analytical History', skiprows = lambda x: x not in rows_to_import_from_raw_data_3)

data2 = data2.rename(columns={'ISO-alpha3 Code': 'Country Code'})
data3 = data3.rename(columns={'Unnamed: 0': 'Country Code'})
data3 = data3[['Country Code', 'Unnamed: 23']]

recession_data_1 = data1.merge(data2, on ='Country Code').merge(data3, on = 'Country Code')
recession_data_1 = recession_data_1.rename(columns = {'Unnamed: 23': 'Income Group'})
gdp_years = [str(i) + ' [YR' + str(i) + ']' for i in range(2006,2012)]
good_name_gdp_years = [str(i) for i in range(2006,2012)]
recession_data_1 = recession_data_1[['Country Code', 'Country Name', 'Region Name', 'Income Group'] + gdp_years]
recession_data_1.columns = ['Country Code', 'Country Name', 'Region Name', 'Income Group'] + good_name_gdp_years

recession_data_1 = recession_data_1.set_index('Country Code')

full_names_of_income_group = {'H' : 'High', 'UM' : 'Upper middle', 'LM' : 'Lower middle', 'L' : 'Low'}
recession_data_1 = recession_data_1.replace({'Income Group' : full_names_of_income_group})

recession_data_1 = recession_data_1[~ recession_data_1.isin(['..', 'Libya']).any(axis = 1)]

recession_data_1[good_name_gdp_years] = recession_data_1[good_name_gdp_years].astype(float).round(2)
recession_data_1
```
